```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('gpascore.csv')
print(data)
```

판다스에서 빈칸을 알아서 세줍니다. gre열에 하나가 비어있어요  
>data.fillna(100)   

빈칸에 100을 넣어주는 함수 입니다.

```python
#print(data.isnull().sum())
data=data.dropna()
#exit()
```

>print(data['gre'].min())

이 열의 데이터가 몇개인지 알려줍니다.

```python
print(data['gpa'])
```

y데이터의 값들을 list 형태로 가져와 줍니다.

```python
yData=data['admit'].values 
print(yData)
```

dataFrame에 붙일수있는 iterrows()는 반복문을 돌리면 데이터를 한행씩 출력할 수 있는 것입니다. 그리고 한 행인 rows안의 gre열에 있는것을 출력해 주는 것입니다.

```python
xData=[]
for i,rows in data.iterrows():
    xData.append([rows['gre'],rows['gpa'],rows['rank']])
```

Dense라는 함수에 넣을 두번째 파라미터는 Activation 입니다.
fix의 첫번째 파라미터는 x데이터(학습데이터), y데이터(정답)가 들어가게 됩니다. 그리고 파이썬 리스트 그대로 넣으면 에러가 뜨기 때문에 numpy array 형식 또는 tf tensor 형식으로 넣어주어야 합니다.

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(64,activation='tanh'),
    tf.keras.layers.Dense(128,activation='tanh'),
    tf.keras.layers.Dense(1,activation='sigmoid'),
]) #딥러닝 모델 만들기

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.fit(np.array(xData),np.array(yData),epochs=1000)
```

만들어진 학습 모델로 예측을 해 보겠습니다. predict는 예측을 해주는 고마운 함수 입니다.

```python
predict=model.predict([[750,3.70,3],[400,2.2,1]])
print(predict)
```
