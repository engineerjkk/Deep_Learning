```python
import tensorflow as tf

Pmodel=tf.keras.models.load_model('drive/My Drive/Colab Notebooks/model1')
```

```python
text=open('drive/My Drive/Colab Notebooks/pianoabc.txt').read()
```

```python
unique_text=list(set(text)) #중복허용하지 않는 리스트입니다.
unique_text.sort() 
```

```python
text_to_num={}
num_to_text={}

for i,data in enumerate(unique_text):
  num_to_text[i]=data
for i,data in enumerate(unique_text):
  text_to_num[data]=i
```

```python
NumText=[]

for i in text:
  NumText.append(text_to_num[i])

print(NumText)
```

```python
FirstValue=NumText[117:117+25]
FirstValue=tf.one_hot(FirstValue,31)
print(FirstValue)
```

>PredictValue=Pmodel.predict(FirstValue)

그냥 해주면 에러가 뜹니다. 학습할때 집어넣은 shape은 (29만,25,31) 이지만 지금은 (25,31) 이기 때문이죠. 따라서 (1,25,31) 해주면되므로 겉에 괄호하나 더 칩니다.

```python
FirstValue=tf.expand_dims(FirstValue,axis=0)
```

```python
print(FirstValue)
```

```python
PredictValue=Pmodel.predict(FirstValue)
```

좀 많이 작은 숫자가 출력되는 군요. 이것은 31개의 확률입니다. 31개중 어디에 있는게 높은지 출력하는 것입니다. 그래서 최대값을 가져옵니다.

```python
print(PredictValue)
```

5번째 문자라고 예측해주는군요.

```python
import numpy as np
np.argmax(PredictValue[0])
```

다섯번째 문자의 값은 2군요 !

```python
print(num_to_text[np.argmax(PredictValue[0])])
```

```python
print(num_to_text[NumText[117+25]])
```

이제 저희가 해야할 것은 
1. 첫 입력값 만들기
2. Predict로 다음 문자 예측하기
3. 예측한 다음문자 [] 저장하기 입니다.
4. 첫입력값 앞에 짜르기
5. 예측한 다음문자를 뒤에 넣기
6. 원핫 인코딩하기, expand dims
6. 그 새로운 입력값을 Predict 하기 x 반복


```python
music=[]
for i in range(200):  
  PredictValue=Pmodel.predict(FirstValue)
  PredictValue=np.argmax(PredictValue[0])

  #음악의 반복이 될 수 있으므로 랜덤도 넣어줍니다.만약 0.9가 나오면 90%확률로 가장 확률이 큰값을 뽑아줍니다.
  #newValue=np.random.choice(unique_text,1,p=PredictValue[0])

  music.append(PredictValue)

  NextInput=FirstValue.numpy()[0][1:]
 # print(NextInput)

  one_hot_num=tf.one_hot(PredictValue,31)
  #print('원핫한거',one_hot_num)

  FirstValue=np.vstack([NextInput,one_hot_num.numpy()])
  FirstValue=tf.expand_dims(FirstValue,axis=0)
```

리스트 두개를 합쳐봅시다  
>np.vstack()

```python
print(music)
```

음악이 작곡됐습니다. ! 이제 반복문을 숫자화된 예측결과를 문자로 바꾸어줍니다.

```python
music_text=[]

for i in music:
  music_text.append(num_to_text[i])
print(music_text)
```
