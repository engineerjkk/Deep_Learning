```python
import tensorflow as tf
```

우리는 눈으로 봐도 답을 압니다. 아마 x에 2를곱하고 1을 더하면 y가 나오겠죠? 선형회귀1 예제에서는 데이터가 하나밖에 없었지만 이젠 여러개죠. 이걸로 모델을 만들어 보겠습니다.

```python
train_x=[1,2,3,4,5,6,7]
train_y=[3,5,7,9,11,13,15]
```

Weight를 만들어줍니다. 변수 안의 초기값은 보통 randomize로 무작위 값을 집어넣어 줍니다.

```python
a=tf.Variable(0.1)
b=tf.Variable(0.1)
```

learning_rate는 잘 나올때 까지 값을 수정해줘야합니다.

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.1)
```

손실함수 : 모델을 통한 예측값과 실제값의 차이입니다.  
>mean squared error  
>cross entropy  

tf.keras.losses.mse(실제값 List, 예측값 List)를 써주면 텐서플러우에서 리스트라도 알아서 전부 한번에 오차를 계산해줍니다.

```python
def loss_function(a,b):
    predict=train_x*a+b
    return tf.keras.losses.mse(train_y,predict)
```

```python
for i in range(3000):
    opt.minimize(loss_function, var_list=[a,b])
    print(a.numpy(),b.numpy())
```

손실함수를 제대로 만들기 위해 함수를 감싸는 lamda를 사용해 줍니다.

```python
for i in range(3000):
    opt.minimize(lambda:loss_function(a,b), var_list=[a,b])
    print(a.numpy(),b.numpy())
```
