```python
text=open('drive/My Drive/Colab Notebooks/pianoabc.txt').read()
print(text)
```

이제 문자를 숫자로 바꾸어야 합니다. 여기 데이터안에 있는 숫자도 문자입니다. 명심하세요. Bag of words 단어주머니를 만듭니다.

```python
unique_text=list(set(text)) #중복허용하지 않는 리스트입니다.
unique_text.sort()
print(unique_text)
```

목표는 {'A':1,'B':2,'C':3...}

```python
text_to_num={}
num_to_text={}

for i,data in enumerate(unique_text):
  num_to_text[i]=data
for i,data in enumerate(unique_text):
  text_to_num[data]=i

print(text_to_num)
print(num_to_text)
```

```python
print(text_to_num['A'])
print(num_to_text[14])
```

원본 텍스트를 모두 숫자로 변환하여 줍니다.

```python
NumText=[]

for i in text:
  NumText.append(text_to_num[i])

print(NumText)
```

trainX는 input, trainY는 정답이 되겠죠.  
이걸 반복문으로 쭉 끝까지 갑니다

```python
 X=[]
 Y=[]

 for i in range(0,len(NumText)-25):
   X.append(NumText[i:i+25])#25개의 리스트를 잘라서 가져옵니다. 이게 X데이터1개입니다.
   Y.append(NumText[i+25]) 
   
print(X[0:5])
print(Y[0:5])
```

```python
import numpy as np
print(np.array(X).shape)
print(np.array(Y).shape)
```

원핫인코딩을 수행합니다. X와 unique의 개수를 파라미터로 넣어줍니다.  
결과를 보니 31개의 항목이 25개 들어있네요. 이걸 두개 출력했구요

```python
import tensorflow as tf
X=tf.one_hot(X,31)
Y=tf.one_hot(Y,31)
print(X[0:2])
```

이제 본격적으로 모델에 집어넣어 봅시다.  
activation은 LSTM에서 이미 많이 넣었으니 따로 안넣어도 됩니다.  
LSTM은 epochs를 많이 해줘야 합니다.  
코랩에서는 epoch길게할때 verbose=2 는 필수입니다. 학습 중 출력되는 것을 줄여주죠 verbose 덕분에 화살표도 안나와요. 그래도 accuracy는 나와야하니 metrics는 추가해 줍시다 !

```python
model=tf.keras.models.Sequential([
  tf.keras.layers.LSTM(100,input_shape=(25,31)), #노드수는 100적어보고, Summary를 위해 input shape을 집어넣죠.
  tf.keras.layers.Dense(31,activation='softmax')#마지막 데이터를 어떻게 예측할까, 31개중 예측하므로, 
])#loss를 categorical_crossentropy를 쓰려면 꼭 softmax를 써줘야합니다.
#데이터가 이미 원핫인코딩이 되어있으면 sparse를 안써도됩니다. 만약데이터가 정수화되어있으면 sparse를 써 주어야겠죠.
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X,Y,batch_size=64,epochs=10)#, verbose=2
model.save('drive/My Drive/Colab Notebooks/model1')
```

이번엔 모델을 만들 때 레이어를 LSTM, Dense 이렇게 하나만 넣어보도록 합시다. 

– LSTM을 두개 세개 중첩해서 연달아 사용할 수 있는데 그럴라면 LSTM( return_sequences=True ) 이거가 필요합니다. 

– 31개의 문자 중 어디 속할지 예측하는 카테고리 예측문제기 때문에 categorical_crossentropy loss 함수가 필요합니다.

– 원핫인코딩 되어있을 때는 sparse_categorical_crossentropy 를 쓰지 않고 저걸 씁니다.

– 마지막 Dense 레이어의 activation 함수는 역시 softmax입니다.

 

– model.fit()할 때 X, Y 집어넣으시면 되고 batch_size=64, shuffle=True 등 자유롭게 설정가능합니다.

– validation 데이터셋은 굳이 필요없을 것 같아서 뺐습니다. 

– Google Colab에서 epoch 30번 이상 많이 돌릴 땐 verbose=2 이게 꼭 필요합니다. 안그러면 다운되고 멈출 수 있음
